```python
import re

# Function to extract min, max, and average experience from the column
def parse_experience(value):
    value = str(value).lower()  # Ensure consistent case
    if pd.isna(value) or 'nan' in value:
        return (None, None, None)  # Handle missing values
    if '-' in value:  # Range like "3 - 5 Yrs of Exp"
        match = re.findall(r'\d+', value)
        if len(match) == 2:
            min_exp, max_exp = int(match[0]), int(match[1])
            avg_exp = (min_exp + max_exp) / 2
            return (min_exp, max_exp, avg_exp)
    elif '+' in value:  # Single value with "+" like "+3 Yrs of Exp"
        match = re.search(r'\d+', value)
        if match:
            min_exp = int(match.group())
            return (min_exp, None, min_exp)  # No max, avg is min_exp
    else:  # Single value without range like "5 Yrs of Exp"
        match = re.search(r'\d+', value)
        if match:
            min_exp = int(match.group())
            return (min_exp, min_exp, min_exp)
    return (None, None, None)  # Fallback for unmatched patterns

# Apply the function to the dataset
df_no_NotSpecified_yrs = Data[Data['Experience Years'] != 'No Specific Yrs Of Exp']
df = pd.DataFrame({'Experience Years': df_no_NotSpecified_yrs['Experience Years']})
df[['Min Exp', 'Max Exp', 'Avg Exp']] = df_no_NotSpecified['Experience Years'].apply(parse_experience).apply(pd.Series)

# Preview the results
print(df[['Experience Years', 'Min Exp', 'Max Exp', 'Avg Exp']].head())
df.to_excel('ExperienceYears.xlsx')
```

```python
# Group the data by 'Government' and 'Job Title' and count occurrences
job_counts_by_government = Data.groupby(['Government', 'Job Title']).size().reset_index(name='Job Count')

# Sort the counts to find the most common job titles within each government
job_counts_by_government = job_counts_by_government.sort_values(by=['Government', 'Job Count'], ascending=[True, False])


# Function to plot top n jobs for each government
def plot_top_jobs_per_government(n=5): # Default to top 5 jobs
    for gov in job_counts_by_government['Government'].unique():
        gov_data = job_counts_by_government[job_counts_by_government['Government'] == gov].head(n)

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Job Count', y='Job Title', data=gov_data, palette=palette)
        plt.title(f'Top {n} Job Titles in {gov} Government')
        plt.xlabel('Number of Jobs')
        plt.ylabel('Job Title')
        plt.show()

plot_top_jobs_per_government() # plot the top 5 jobs per government
```

```python
from collections import Counter
def plot_common_skills(top_n=10):

    # Aggregate skill counts per job title
    if 'Skills' not in Data.columns:
        print("Error: 'Skills' column not found in the DataFrame. Please ensure it exists.")
        return

    skills_by_job = Data.groupby('Job Title')['Skills'].apply(lambda x: ' '.join(x)).reset_index()

    top_job_titles = Data['Job Title'].value_counts().nlargest(top_n).index

    skills_by_job = skills_by_job[skills_by_job['Job Title'].isin(top_job_titles)]

    all_skills = ' '.join(skills_by_job['Skills']).split()
    top_skills = dict(Counter(all_skills).most_common(5))
    plt.figure(figsize=(12, 6))
    sns.barplot(x=list(top_skills.values()), y=list(top_skills.keys()), palette=palette)
    plt.title('Top 5 Skills for top 10 Jobs')
    plt.xlabel('Skill Count')
    plt.ylabel('Skill')
    plt.show()

# Call the function
plot_common_skills()
```

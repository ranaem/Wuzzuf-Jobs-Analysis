# **Data Preparation & Cleaning**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Merging the Data I scraped from wuzzuf and that I grabbed from Kaggle

```python
excel_files = [
    "Wuzzuf1.xlsx",
    "Wuzzuf2.xlsx",
    "Wuzzuf3.xlsx",
    "Wuzzuf4.xlsx",
    "Wuzzuf5.xlsx",
    "Wuzzuf6.xlsx",
    "Wuzzuf7.xlsx",
    "Wuzzuf8.xlsx",
    "Wuzzuf-Jobs.xls"
]

dataframe = []

for file in excel_files:
    df = pd.read_excel(file)
    dataframe.append(df)

merged_data = pd.concat(dataframe, ignore_index=True)

merged_data.to_excel('WuzzufJobs.xlsx', index = False)
```

Due to scrapping the Skills column had records for not only skills so I split it to (Job Type, On-site/Remote, Experience Level, Experience Years and the skills)

```python
file_path = "WuzzufJobs.xlsx"


df = pd.read_excel(file_path)


df['Job Type'] = df['Skills'].str.findall(r'(Full Time|Part Time|Freelance|Internship|Remote)').apply(lambda x: ', '.join(x) if x else None)


df['On-site/Remote'] = df['Skills'].str.findall(r'(On-site|Remote|Hybrid)').apply(lambda x: ', '.join(x) if x else None)


df['Experience Level'] = df['Skills'].str.findall(r'(Experienced|Entry Level|Manager|Student|Senior Level)').apply(lambda x: ', '.join(x) if x else None)


df['Experience Years'] = df['Skills'].str.extract(r'(\d+ - \d+ Yrs of Exp|\d+\+ Yrs of Exp)')

df['Skills Only'] = df['Skills'].str.replace(
    r'(Full Time|Part Time|Freelance|Internship|Remote|On-site|Hybrid|Experienced|Entry Level|Manager|Student|Senior Level|\d+ - \d+ Yrs of Exp|\d+\+ Yrs of Exp|·)',
    '',
    regex=True
).str.strip()


output_file = "WuzzufJobs_Cleaned.xlsx"
df.to_excel(output_file, index=False)

print("تم تقسيم البيانات وحفظها في ملف جديد.")
```

Then I started to Prepare my Data

```python
df = pd.read_excel("WuzzufJobs_Cleaned.xlsx")

#Dropping the original skills column and date because I don't need it
df.drop(columns=['Skills'], inplace=True)
df.drop(columns=['Date'], inplace=True)
df.to_excel("WuzzufJobs_Cleaned.xlsx", index=False)
```

```python
df = df.rename(columns={'Skills Only': 'Skills'})
df.to_excel("WuzzufJobs_Cleaned.xlsx", index=False)
```

```python
df.head()
```

```python
#stripping the '-' in the end of each company
df['Company'] = df['Company'].str.replace(' -', '')
df.to_excel("WuzzufJobs_Cleaned.xlsx", index=False)
```

```python
df.head()
```

```python
#Modify Skills column
df['Skills'] = df['Skills'].astype(str).str.lstrip('/')

df.to_excel("WuzzufJobs_Cleaned.xlsx", index=False)
df.head()
```

```python
#Dropping Link column
df.drop('Link', axis=1, inplace=True)
df.to_excel("WuzzufJobs_Cleaned.xlsx", index=False)
```

```python
df.head()
```

Data Cleaning

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Fill missing values in 'Job Type' with 'Not Specified'
df['Job Type'].fillna('Not Specified', inplace=True)

#Fill missing values in 'On-site/Remote' with 'Not Specified'
df['On-site/Remote'].fillna('Not Specified', inplace=True)

# Fill missing values in 'Experience Level' with 'Not Specified'
df['Experience Level'].fillna('Not Specified', inplace=True)

# Fill missing values in 'Experience Years' with 'No Specific Yrs Of Exp'
df['Experience Years'].fillna('No Specific Yrs Of Exp', inplace=True)

# Verify the changes
print(df.isnull().sum())

#Save the updated DataFrame back to the Excel file
df.to_excel("WuzzufJobs_Cleaned.xlsx", index=False)
```
